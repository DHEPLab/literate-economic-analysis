# Stata Kernel

Code cells as if you are writing in the command mode / do file editor. There are some **limitations** with this interface in that you don't have the `data editor` or other `gui` features available to you. Some of these limitations are addressed through [magics](https://kylebarron.dev/stata_kernel/using_stata_kernel/magics/). Note that Python code won't work with this approach.

```stata
sysuse auto
```

```stata
list rep78 in 46/56
```

## Magics

```stata
%browse
```

You can select just the first `10 observations` using `%head`

```stata
%head
```

```stata
%tail
```

### Some other basic stats

```stata
tabulate foreign headroom
```

```stata
scatter price weight
```

```stata
reg weight price
```

```stata
scatter weight price || lfit weight price
```

You can access help using `%help`

```stata
%help scatter
```

You're not as constrained as the `command` window in that you can write multiple commands in a single `code-cell`

```stata
list rep78 in 1/2
scatter length weight
```
